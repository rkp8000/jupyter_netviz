```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from animation import animate_network
```

```python
N_NODES = 20  # NUMBER OF NODES IN NETWORK
P_CXN = .1  # PROBABILITY OF CONNECTION BETWEEN NODES
DT = .1  # TIMESTEP BETWEEN DISPLAY UPDATES
N_FRAMES = 10  # NUMBER OF FRAMES
```

```python
# make a random connection matrix
cxns = np.random.rand(N_NODES, N_NODES) < P_CXN
# assign random weights to connections
weights = np.random.rand(N_NODES, N_NODES)
weights[~cxns] = 0
```

```python
# make random activations
activations = np.random.rand(N_FRAMES, N_NODES)
```

```python
# animate the network
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
animate_network(ax, weights, activations, DT)
```
